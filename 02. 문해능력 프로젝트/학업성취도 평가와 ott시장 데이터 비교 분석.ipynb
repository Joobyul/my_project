### OECD회원국 학업성취도평가 CSV 데이터 불러오기

```python
file='학업_성취도_OECD회원국__20230714022025.xlsx'
```

```python
oecdDF=pd.read_excel(file, header=0)
oecdDF
```

### 국가별 컬럼을 인덱스로 지정

```python
oecdDF.set_index("국가별")
```

```python
oecdDF=oecdDF[0:3]
```

```python
oecdDF
```

### 연도별/성별 읽기점수 데이터 생성

```python
oecdDF=oecdDF.loc[:, ['2012.4', '2012.5', '2015.4', '2015.5', '2018.4', '2018.5']]
```

```python
oecdDF
```

```python
oecdDF=oecdDF.drop(1)
oecdDF
```

### 연도별 읽기점수 데이터 생성

```python
oecdDF=oecdDF.drop(0)
oecdDF
```

```python
oecdDf=oecdDF.astype(int)
```

```python
oecdDF.mean()
```

```python
df_means = oecdDF.groupby(oecdDF.columns, axis=1).mean()
```

```python
new_data = {'연도': df_means.columns,
            '읽기점수': df_means.values.flatten()}

df_combined = pd.DataFrame(new_data)
```

```python
df_combined
```

```python
df_means = df_combined.groupby(df_combined['연도'].str.extract(r'(\d{4})')[0])['읽기점수'].mean().reset_index()
```

```python
oecdDF=df_means.set_index(0)
```

```python
oecdDF
```

```python
ottDF=pd.read_excel(file1)
```

```python
ottDF=ottDF.fillna(method='bfill')
ottDF
```

### NaN 값을 앞의 값으로 대체하여 채우기

### 필요없는 컬럼 제거

```python
ottDF
```

### OECD 회원국 읽기점수 연도만 발췌한 데이터 생성

### 시각화(OECD회원국 읽기점수와 OTT 시장변화)

```python
plt.style.use('_mpl-gallery')
fig=plt.figure(figsize=(10, 10))
ax1=fig.add_subplot(2,1,1)
ax1.fill_between(oecdDF.index, oecdDF['읽기점수'], alpha=.5, linewidth=0)
ax1.plot(oecdDF, marker='o', markerfacecolor='green', markersize=10, 
        color='olive', linewidth=2, label='읽기점수')
ax1.set(ylim=(500,550))
ax1.legend(loc='best')
ax1.set_xlabel('연도')
ax1.set_ylabel('평가점수')
ax2=fig.add_subplot(2,1,2)
ax2.plot(ottDF['구독자 수(단위:백만)'], marker='o', markerfacecolor='skyblue', markersize=10, 
        color='blue', linewidth=2, label='구독자 수')
ax2.plot(ottDF['콘텐츠 제작비용(단위:억)'], marker='o', markerfacecolor='green', markersize=10, 
        color='olive', linewidth=2, label='제작비용')
ax2.plot(ottDF['시청시간'], marker='o', markerfacecolor='red', markersize=10, 
        color='magenta', linewidth=2, label='시청시간')
ax2.legend(loc='best')
ax2.set_xlabel('연도')
ax2.set_ylabel('ott 시장변화')
```

```python
plt.style.use('ggplot')


df_3=ottDF.loc[ottDF.index, ['구독자 수(단위:백만)', '콘텐츠 제작비용(단위:억)', '시청시간']]
ax1=df_3.plot(kind='bar',figsize=(10,5), width=0.7, color=['orange', 'green', 'skyblue'])
ax2=ax1.twinx()
ax2.plot(oecdDF, marker='o', markerfacecolor='red', markersize=10, 
        color='magenta', linewidth=2, label='읽기점수')

ax2.set_ylim(510,550)

ax1.legend(loc='best')
ax2.legend(loc='best')

ax2.set_ylabel('독서율')
ax1.set_xlabel('연도')
ax1.set_ylabel('ott 시장변화')

plt.savefig('ott시장변화.png')
```
